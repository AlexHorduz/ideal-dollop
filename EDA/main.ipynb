```python
import os

import matplotlib.pyplot as plt

from helpers import (
    read_labels, count_labels, count_labels_per_image, show_image_grid,
)

BASE_DATA_PATH = "../dataset/car"

LABELS = [
    "Green Light", "Red Light", "Speed Limit 10", "Speed Limit 100",
    "Speed Limit 110", "Speed Limit 120", "Speed Limit 20", "Speed Limit 30",
    "Speed Limit 40", "Speed Limit 50", "Speed Limit 60", "Speed Limit 70",
    "Speed Limit 80", "Speed Limit 90", "Stop",
]
```

```python
split_names = ["train", "valid", "test"]

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

for ax, split_name in zip(axes, split_names):
    labels = read_labels(os.path.join(BASE_DATA_PATH, split_name, "labels"))
    counts = count_labels(labels)

    # # Convert counts to label names
    x = LABELS
    y = [counts.get(i, 0) for i in range(len(LABELS))]

    ax.bar(x, y)
    ax.set_xlabel("Label name")
    ax.set_ylabel("Number of instances")
    ax.set_title(f"{split_name} dataset (total: {len(labels)})")
    ax.tick_params(axis="x", rotation=90)
plt.suptitle("Number of instances for each class")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)

for ax, split_name in zip(axes, split_names):
    labels = read_labels(os.path.join(BASE_DATA_PATH, split_name, "labels"))
    counts_per_image = count_labels_per_image(labels)

    ax.hist(counts_per_image, bins=range(1, max(counts_per_image)+2), edgecolor="black", density=True)
    ax.set_xlabel("Number of labels per image")
    ax.set_ylabel("Number of images (normalized)")
    ax.set_title(f"{split_name} dataset (images: {len(labels)})")

plt.suptitle("Number of objects per image")
plt.tight_layout()
plt.show()
```
